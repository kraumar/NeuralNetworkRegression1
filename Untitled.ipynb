# 1. Loading the dataset.

```python
import pandas as pd

df = pd.read_csv('Bias_correction_ucl.csv')
df.describe()
```

```python
df.head()
```

# 2. Data preprocessing

2.1 Deleting rows with missing values

```python
import numpy as np

def delete_missing(df, col_names): # (df, ['arg1', ...., 'argn'])
    new_df_copy = df.copy()
    #mark with NaN value
    new_df_copy = new_df_copy[col_names].replace('?', np.NaN)
    #remowe rows and return
    return new_df_copy.dropna()



df_list_col = list(df.columns.values)
#deleting date col
df_list_col.remove('Date')
#deleting missing
df = delete_missing(df,df_list_col)
df.describe()
```

2.2 One hot encoding

```python
def oneHotEncode(df,colNames):
    for col in colNames:
        if( df[col].dtype == np.dtype('object')):
            dummies = pd.get_dummies(df[col],prefix=col)
            df = pd.concat([df,dummies],axis=1)

            #drop the encoded column
            df.drop([col],axis = 1 , inplace=True)
    return df



print('There were {} columns before encoding categorical features'.format(df.shape[1]))
df = oneHotEncode(df,df_list_col)
print('There are {} columns after encoding categorical features'.format(df.shape[1]))
```

There were no non-numerical cols

2.3 X/y1/y2 split

```python

list_noY1Y2 = df_list_col
list_noY1Y2.remove('Next_Tmax')
list_noY1Y2.remove('Next_Tmin')

X = df[list_noY1Y2]
y1 = df['Next_Tmax']
y2 = df['Next_Tmin']
```

2.4 Train/Test split

```python
from sklearn.model_selection import train_test_split

X1_train, X1_test, y1_train, y1_test = train_test_split(X, y1, test_size=0.33, random_state=42)
X2_train, X2_test, y2_train, y2_test = train_test_split(X, y2, test_size=0.33, random_state=42)


```

```python
print('Training data 1 shape : ', X1_train.shape, y1_train.shape)
print('Testing data 1 shape : ', X1_test.shape, y1_test.shape)

print('Training data 2 shape : ', X1_train.shape, y1_train.shape)
print('Testing data 2 shape : ', X1_test.shape, y1_test.shape)
```

# 3. Building Network

```python
from keras import models
from keras import layers

model1f = models.Sequential()
model1f.add(layers.Dense(64, activation='relu',
                           input_shape=(X1_train.shape[1],)))
model1f.add(layers.Dense(64, activation='relu'))
model1f.add(layers.Dense(1))
model1f.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])

model2f = models.Sequential()
model2f.add(layers.Dense(64, activation='relu',
                           input_shape=(X2_train.shape[1],)))
model2f.add(layers.Dense(64, activation='relu'))
model2f.add(layers.Dense(1))
model2f.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
```

3.2 GridSearch

```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
import sklearn

def algorithm_pipeline(X_train_data, X_test_data, y_train_data, y_test_data, 
                       model, param_grid, cv=4, scoring_fit='neg_mean_absolute_error',
                       do_probabilities = False):
    gs = GridSearchCV(
        estimator=model,
        param_grid=param_grid, 
        cv=cv, 
        n_jobs=-1, 
        scoring=scoring_fit,
        verbose=2
    )
    fitted_model = gs.fit(X_train_data, y_train_data)
    
    if do_probabilities:
      pred = fitted_model.predict_proba(X_test_data)
    else:
      pred = fitted_model.predict(X_test_data)
    
    return fitted_model, pred

def build_nn1():
    model = models.Sequential()
    model.add(layers.Dense(64, activation='relu',
                           input_shape=(X1_train.shape[1],)))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    
    return model

def build_nn2():
    model = models.Sequential()
    model.add(layers.Dense(64, activation='relu',
                           input_shape=(X1_train.shape[1],)))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    
    return model

param_grid = {
              'epochs':np.arange(1,20,1),
              }

model1 = KerasClassifier(build_fn = build_nn1, verbose=0)
model1, pred1 = algorithm_pipeline(X1_train, X1_test, y1_train, y1_test, model1, 
                                        param_grid)

model2 = KerasClassifier(build_fn = build_nn2, verbose=0)
model2, pred2 = algorithm_pipeline(X2_train, X2_test, y2_train, y2_test, model2, 
                                        param_grid)
```

```python
print("model 1:")
print(model1.best_score_)
print(model1.best_params_)
print("")
print("model 2: ")
print(model2.best_score_)
print(model2.best_params_)
```
